```python
statColumns = ['Top Blue Kill','Jungle Blue Kill','Mid Blue Kill','ADC Blue Kill','Support Blue Kill',
               'Top Blue Death','Jungle Blue Death','Mid Blue Death','ADC Blue Death','Support Blue Death',
               'Top Blue Assist','Jungle Blue Assist','Mid Blue Assist','ADC Blue Assist','Support Blue Assist',
               'Top Red Kill','Jungle Red Kill','Mid Red Kill','ADC Red Kill','Support Red Kill',
               'Top Red Death','Jungle Red Death','Mid Red Death','ADC Red Death','Support Red Death',
               'Top Red Assist','Jungle Red Assist','Mid Red Assist','ADC Red Assist','Support Red Assist'
              ]
zeromatrix = pd.DataFrame(np.zeros((190060,30)),columns = statColumns)
killsData = pd.concat([killsData,zeromatrix],axis = 1)
killsData.head()
```

```python
#one off replacements
killsData.loc[:,"Victim":"Assist_4"] = killsData.loc[:,"Victim":"Assist_4"].apply(lambda x: x.astype(str).str.lower())
killsData.loc[:,"Top Blue Tag":"Support Red Tag"] = killsData.loc[:,"Top Blue Tag":"Support Red Tag"].apply(lambda x: x.astype(str).str.lower())
killsData = killsData.replace('najin','nje',regex = True)
killsData = killsData.replace('sbenu','ssb',regex = True)
killsData = killsData.replace('  ',' ',regex = True)
killsData = killsData.replace('raison','hke raison')
killsData = killsData.replace('lms','m17',regex = True)
killsData = killsData.replace(' $','',regex = True)
killsData = killsData.replace('^ ','',regex = True)
```

```python
killsData = killsData.replace('m17 m17dee','m17dee')
killsData = killsData.replace('m17dee','m17 dee')
```

```python
killsData = killsData.replace('^19 ','m19 ', regex = True) 
killsData = killsData.replace('^v ','vp ', regex = True) 
```

```python
#Apply counters
for i, row in killsData.iterrows():
    if row['Victim'] == 'nan':
        continue
    else:
        statDict = {
            row['Top Blue Tag']: 'Top Blue',
            row['Jungle Blue Tag']: 'Jungle Blue',
            row['Mid Blue Tag']: 'Mid Blue',
            row['ADC Blue Tag']: 'ADC Blue',
            row['Support Blue Tag']: 'Support Blue',
            row['Top Red Tag']: 'Top Red',
            row['Jungle Red Tag']: 'Jungle Red',
            row['Mid Red Tag']: 'Mid Red',
            row['ADC Red Tag']: 'ADC Red',
            row['Support Red Tag']: 'Support Red'
        }

        victimPos = statDict[row['Victim']]
        killerPos = statDict[row['Killer']]
        assistList = [row['Assist_1'],row['Assist_2'],row['Assist_3'],row['Assist_4']]
        cleanedAssistList = [x for x in assistList if str(x) != 'nan']
        assistPos = []
        for x in cleanedAssistList:
            assistPos.append(statDict[x])

        death = '{} Death'.format(victimPos)
        killsData.at[i,death] = 1
        kill = '{} Kill'.format(killerPos)
        killsData.at[i,kill] = 1
        for j in assistPos:
            assist = '{} Assist'.format(j)
            killsData.at[i,assist] = 1
```

```python
#Create Time Buckets
def timebucket(row):
    if row['Time'] >=1 and row['Time'] <5:
        return('1 - 5 minute mark')
    elif row['Time'] >=5 and row['Time'] <10:
        return('5 - 10 minute mark')
    elif row['Time'] >=10 and row['Time'] <15:
        return('10 - 15 minute mark')
    else:
        return('15 - 20 minute mark')
```
